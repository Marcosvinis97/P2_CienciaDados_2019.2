## Preparando o ambiente no jupyter:

```python
import tweepy
import math
import os.path
import pandas as pd
import json
from random import shuffle
```

___
## Autenticando no  Twitter


```python
#leitura do arquivo no formato JSON
with open('auth.pass') as fp:    
    data = json.load(fp)

#Configurando a biblioteca. N√£o modificar
auth = tweepy.OAuthHandler(data['consumer_key'], data['consumer_secret'])
auth.set_access_token(data['access_token'], data['access_token_secret'])
```

___
## Etapas do projeto:

### Escolha de um produto e coleta das mensagens


```python
#Produto escolhido:
produto = 'depressao'

#Quantidade m√≠nima de mensagens capturadas:
n = 250
#Quantidade m√≠nima de mensagens para a base de treinamento:
t = 450

#Filtro de l√≠ngua, escolha uma na tabela ISO 639-1.
lang = 'pt'
```

Capturando os dados do twitter:

```python
msgs = []
```

```python
planilha = pd.read_excel('depressao.xlsx', "Treinamento")
```

```python
#Cria um objeto para a captura
api = tweepy.API(auth)
#Inicia a captura, para mais detalhes: ver a documenta√ß√£o do tweepy
i = 1
for msg in tweepy.Cursor(api.search, q="{0} -filter:retweets".format(produto), lang=lang, tweet_mode="extended").items():
    new_msg = { 'User Name': msg.user.name,
                'Tweet Created At': msg.created_at,
                'Tweet Text': msg.full_text.lower(),
                'User Location': msg.user.location,
                'Phone Type': msg.source,
                'Favorite Count': msg.favorite_count,
                'Retweets':msg.retweet_count
                }
    if new_msg["Tweet Text"] not in planilha["Tweet Text"]:
        msgs.append(new_msg)
    i += 1
    if i > n:
        break

#Embaralhando as mensagens para reduzir um poss√≠vel vi√©s
# shuffle(msgs)
```

```python
pd.set_option('display.max_colwidth', -1)
pd.DataFrame(msgs)
```

Salvando os dados em uma planilha Excel:

```python
#Verifica se o arquivo n√£o existe para n√£o substituir um conjunto pronto
if not os.path.isfile('./{0}.xlsx'.format(produto)):
    
    #Abre o arquivo para escrita
    writer = pd.ExcelWriter('{0}.xlsx'.format(produto))

    #divide o conjunto de mensagens em duas planilhas
    dft = pd.DataFrame(msgs[:t])
    dft.to_excel(excel_writer = writer, sheet_name = 'Treinamento', index = False)

    dfc = pd.DataFrame(msgs[t:])
    dfc.to_excel(excel_writer = writer, sheet_name = 'Teste', index = False)

    #fecha o arquivo
    writer.save()
```
